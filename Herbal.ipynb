```python
from pypdf import PdfReader
import os
files = []

for file in os.listdir('data'):
    if file.lower().endswith('.pdf'):
        files.append(os.path.join('data',file))

docs = []

for file in files:
    loader = PyPDFLoader(file)
    loaded_docs = loader.load()
    docs.extend(loaded_docs)
```

#### 4. Test spliting

- **CharacterTextSplitter** — Splits text into fixed-size chunks based purely on character count.
- **RecursiveCharacterTextSplitter** — Recursively splits text using separators (paragraph → sentence → word) to preserve context.
- **TokenTextSplitter** — Splits text based on token count to match LLM input limits.
- **NLTKTextSplitter** — Splits text into chunks using sentence boundaries via NLTK.
- **SpacyTextSplitte**r — Uses SpaCy NLP to split text into linguistically meaningful sentences.
- **MarkdownHeaderTextSplitter** — Splits text based on Markdown headers to preserve document structure.
- **HTMLHeaderTextSplitter** — Divides text using HTML tags and structure.
- **SemanticChunker** — Splits text based on semantic meaning using embeddings similarity.
- **SentenceTransformersTokenTextSplitte**r — Combines sentence-transformers with token limits for better semantic chunking.
- **LatexTextSplitter** — Splits LaTeX documents based on sections and formatting.

```python
# creating the text spliter
text_spliter = RecursiveCharacterTextSplitter(
    chunk_size = 1500,
    chunk_overlap = 300,
    separators=["\n\n", "\n", ".", " ", ""]
)

pdf_chuncks = text_spliter.split_documents(docs)

print(pdf_chuncks[0].page_content)
```

### 13. Evualation with LLM(LLM as a Judge)

##### 1. RAG ANSWER (SINGLE SOURCE)

```python
# ===== FINAL RAG ANSWER =====
question = 'uses of Punarnava'
response = rag_chain.invoke({"input": question})
answer = response["answer"]

print("\n===== FINAL ANSWER =====\n")
print(answer)
```

##### 2. RAG SYSTEM EVALUATION (LLM JUDGE)

```python
# ===== RAG SYSTEM EVALUATION =====

retrieved_docs = retriver.invoke(question)
context = "\n\n".join([doc.page_content for doc in retrieved_docs])

eval_prompt_rag = f"""
You are evaluating a RAG system.

Question:
{question}

Retrieved Context:
{context}

Generated Answer:
{answer}

Evaluate system performance:

1. Context Precision (0-1):
Are retrieved docs relevant?

2. Context Recall (0-1):
Did we retrieve enough info?

3. Faithfulness (0-1):
Is answer grounded in context?

Rules:
- Be strict
- Penalize hallucination
- Penalize missing info

Return ONLY JSON:
{{
  "context_precision": float,
  "context_recall": float,
  "faithfulness": float
}}
"""

eval_rag = groq.invoke(eval_prompt_rag)

print("\n===== RAG SYSTEM EVALUATION =====\n")
print(eval_rag.content)
```

##### 3. PARSE RAG EVALUATION

```python
# ===== PARSE RAG EVALUATION =====

import json, re

try:
    json_str_rag = re.search(r"\{.*\}", eval_rag.content, re.DOTALL).group()
    rag_scores = json.loads(json_str_rag)
except:
    rag_scores = {
        "context_precision": 0,
        "context_recall": 0,
        "faithfulness": 0
    }

print("Parsed RAG Scores:", rag_scores)
```

##### 4. ANSWER EVALUATION (LLM JUDGE)

```python
# ===== ANSWER EVALUATION =====

eval_prompt_llm = f"""
You are evaluating ONLY the answer quality.

Question:
{question}

Answer:
{answer}

Evaluate:

1. Answer Relevance (0-1):
Does it answer the question?

2. Completeness (0-1):
Is the answer complete?

3. Clarity (0-1):
Is it well-structured and clear?

Rules:
- Ignore retrieval
- Focus ONLY on answer quality

Return ONLY JSON:
{{
  "answer_relevance": float,
  "completeness": float,
  "clarity": float
}}
"""

eval_llm = groq.invoke(eval_prompt_llm)

print("\n===== ANSWER EVALUATION =====\n")
print(eval_llm.content)
```

##### 5. ANSWER EVALUATION (LLM JUDGE)

```python
# ===== ANSWER EVALUATION =====

eval_prompt_llm = f"""
You are evaluating ONLY the answer quality.

Question:
{question}

Answer:
{answer}

Evaluate:

1. Answer Relevance (0-1):
Does it answer the question?

2. Completeness (0-1):
Is the answer complete?

3. Clarity (0-1):
Is it well-structured and clear?

Rules:
- Ignore retrieval
- Focus ONLY on answer quality

Return ONLY JSON:
{{
  "answer_relevance": float,
  "completeness": float,
  "clarity": float
}}
"""

eval_llm = groq.invoke(eval_prompt_llm)

print("\n===== ANSWER EVALUATION =====\n")
print(eval_llm.content)
```

```python
# ===== FINAL SCORES =====

# ===== PARSE ANSWER EVALUATION =====

import json, re

try:
    json_str_llm = re.search(r"\{.*\}", eval_llm.content, re.DOTALL).group()
    answer_scores = json.loads(json_str_llm)
except:
    answer_scores = {
        "answer_relevance": 0,
        "completeness": 0,
        "clarity": 0
    }

print("\nParsed Answer Scores:", answer_scores)

rag_score = (
    rag_scores["context_precision"] +
    rag_scores["context_recall"] +
    rag_scores["faithfulness"]
) / 3

answer_score = (
    answer_scores["answer_relevance"] +
    answer_scores["completeness"] +
    answer_scores["clarity"]
) / 3

print("\n===== FINAL SUMMARY =====")
print("RAG Score        :", round(rag_score, 3))
print("Answer Score     :", round(answer_score, 3))
```
